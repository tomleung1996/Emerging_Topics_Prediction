```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load the numpy array with shape (2643, 17, 11)
# (term_num, time_steps, feature_num)

gene_arr = pickle.load(open('output/gene_editing/result.array', mode='rb'))
trans_arr = pickle.load(open('output/transplant/result.array', mode='rb'))

gene_terms = pickle.load(open('output/gene_editing/ordered_list.list', mode='rb'))
trans_terms = pickle.load(open('output/transplant/ordered_list.list', mode='rb'))

print('Shape of the gene_editing data:', gene_arr.shape)
print('Shape of the transplant data:', trans_arr.shape)
```

### 文档频率分布情况（不区分年份）

```python
gene_df = np.sum(gene_arr[:, :, 0], axis=1)
trans_df = np.sum(trans_arr[:, :, 0], axis=1)

gene_df_counter = Counter(gene_df)
trans_df_counter = Counter(trans_df)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4), sharex=True, sharey=True)

ax1.hist(gene_df)
ax1.set_title('Gene editing')
ax2.hist(trans_df)
ax2.set_title('Transplant')

ax1.set_yscale('log')
ax2.set_yscale('log')

# add a big axis, hide frame
fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.xlabel("Document frequency")
plt.ylabel("Frequency")

plt.tight_layout()
plt.savefig(r'C:/Users/Tom Leung/Desktop/fig1.png',bbox_inches='tight')
plt.show()
```
