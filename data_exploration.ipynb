```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load the numpy array with shape (2643, 17, 11)
# (term_num, time_steps, feature_num)

arr = pickle.load(open('output/result.array', mode='rb'))
print('Shape of the result array:', arr.shape)
```

```python
count = 0
for i in range(len(arr)):
    flag = False
    
    for j in range(len(arr[0])):
        if arr[i][j][0] == 0:
            flag = True
            break
    
    if not flag:
        count += 1
print('Number of terms that appeared in all years:', count)
```

```python
count = {}
for i in range(len(arr)):
    c = 0
    for j in range(len(arr[0])):
        if arr[i][j][-1] != 0:
            c += 1
    
    count[i] = c
print(Counter(count.values()))
```

```python
print('The annual document frequency change of "clustered regularly interspaced short palindromic repeats:"')
print(arr[97, :, 0].astype(int))

print('The annual Emerging Score change of "clustered regularly interspaced short palindromic repeats:"')
print(arr[97, :, -1])
```

```python
fig, ax = plt.subplots(1, 1)
years = list(range(2003, 2019 + 1))
df_change = arr[:, :, 0]
es_change = arr[:, :, -1]

for i in range(len(arr)):
    ax.plot(years, df_change[i])

plt.show()
```

```python
fig2, ax2 = plt.subplots(1, 1)
for i in range(len(arr)):
    ax2.plot(years, es_change[i])

plt.show()
```
