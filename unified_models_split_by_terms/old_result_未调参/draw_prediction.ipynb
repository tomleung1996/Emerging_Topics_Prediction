```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rnn_dict = {}
with open(r'rnn_gene_top.csv', mode='r', encoding='utf-8') as file:
    for line in file:
        s = line.strip().split(',')
        rnn_dict[s[1]] = np.array(list(map(float, s[2:])))
```

```python
tree_dict = {}
with open(r'tree_gene_top.csv', mode='r', encoding='utf-8') as file:
    for line in file:
        s = line.strip().split(',')
        tree_dict[s[1]] = np.array(list(map(float, s[2:])))
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16, 3))

axes = [ax1, ax2, ax3, ax4]
i = 0
for k, v in rnn_dict.items():
    axes[i].plot(range(5), v, marker='o', markersize=8, label='LSTM')
    axes[i].plot(range(5), tree_dict[k], marker='o', markersize=8, label='LightGBM')
    axes[i].grid(axis='y', ls='--')
    axes[i].legend()
    axes[i].tick_params(axis='both', which='major', labelsize=13)
    axes[i].set_xticks(range(5))
    axes[i].set_xticklabels(['2019', '2020', '2021', '2022', '2023'])
    axes[i].set_title(k, fontsize=15)
    i += 1

ax1.set_ylabel('Popularity score', fontsize=15)
fig.text(0.5, -0.1, 'Forecast years', ha='center', fontsize=18)
plt.tight_layout()
plt.savefig('gene_prediction.jpg', bbox_inches="tight")
plt.show()
```
